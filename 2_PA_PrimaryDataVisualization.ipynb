## Primary Data Visualisation

### Revisiting the description of columns

For the visualisation, only training dataset is used.

```python
train_cleaned.describe(include="all")
```

### Loan Status

As it has only 2 unique values (type object) - bar chart is better to visualise

```python
fig = plt.figure() # create figure

ax0 = fig.add_subplot(2, 1, 1) # add subplot 1 (1 row, 2 columns, first plot)
ax1 = fig.add_subplot(2, 1, 2) # add subplot 2 (1 row, 2 columns, second plot)

train_cleaned["Loan Status"].value_counts().plot(kind="barh", color='steelblue', ax=ax0, figsize=(7, 10))
ax0.set_xlabel('Count - Loan Status')
ax0.set_title('Loan Status distribution')

train_cleaned["Loan Status"].value_counts().plot(kind="pie",
                                                figsize=(7, 10),
                                                autopct='%1.1f%%',
                                                startangle=90,
                                                labels=None,
                                                shadow=True,
                                                pctdistance=1.2,
                                                explode = [0.1, 0],
                                                ax=ax1)

ax1.axis('equal') 
# add legend
ax1.legend(labels=train_cleaned["Loan Status"].unique(), loc='upper left') 

plt.show()
```

### Current loan amount

```python
count, bin_edges = np.histogram(train_cleaned["Current Loan Amount"])

train_cleaned["Current Loan Amount"].plot(kind='hist', figsize=(8, 5), xticks=bin_edges, edgecolor='black', linewidth=1.2, color = "lightcoral")

plt.title('Histogram of current loan amount') # add a title to the histogram
plt.ylabel('Number of Customers') # add y-label
plt.xlabel('Current loan amount') # add x-label

plt.show()
```

## Correlation Matrix

```python
Continuous = ['Loan Status', 'Current Loan Amount', 'Credit Score', 'Annual Income', 'Monthly Debt', 'Years of Credit History', 'Number of Open Accounts', 'Current Credit Balance', 'Maximum Open Credit', 'Bankruptcies', 'Tax Liens', 'Number of Credit Problems']
Category = ['Loan Status', 'Term', 'Years in current job', 'Home Ownership', 'Purpose']

plt.figure(figsize=(12,20))
for i in range(1, 12):
  plt.subplot(6,2,i)
  sns.distplot(train_cleaned.loc[train_cleaned['Loan Status']=='Fully Paid', Continuous[i]], label='Fully Paid', kde_kws={'bw':0.1})
  sns.distplot(train_cleaned.loc[train_cleaned['Loan Status']=='Charged Off', Continuous[i]], label= 'Charged Off', kde_kws={'bw':0.1})
  plt.legend(loc = 'best')
plt.show()
```
